## GIT
### Introducción 

**GIT** es una herramienta conocida como *controlador de versiones*, lo cual hace referencia a una herramienta que nos permite mantener un historial de los cambios hechos a un proyecto. Lo que hace es tomar *snapshots* a lo largo del tiempo sobre el proyecto que estamos trabajando. 

De esta manera se puede conocer:
* Quién hizo las últimas modificaciones al proyecto
* Cuándo se hicieron las modificaciones

### Terminología

Los términos que se utilizan en GIT son:

1. Tree --> Hace referencia a las carpetas o a la raíz de la información.
2. Blob --> Se refiere a los archivos que se encuentran dentro de la raíz (tree).
3. Commit --> Son los *snapshots* que captura Git para su almacenamiento. Conjunto de **blobs** y **trees**.

Es importante decir que la forma en la que Git guarda los *commits* no es de una forma lineal tradicional, sino que es ~~gráfica acíclica directa~~ (**direct acyclical graph**). Esto es contrario a cómo se almacenan los diferentes proyectos, guardando con diferentes versiones dentro de una misma carpeta (*ejercicio version2; ejercicio final; ejercicio final final*).

### Beneficios

El beneficio más grande que muestra dicho programa es el registro a lo largo del tiempo, el cual permite tener conocimiento de los cambios que se hicieron, la **descripción** del por qué se realizó ese cambio y la **fecha y hora** de los cambios realizados. Esto permite tener el conocimiento histórico y en caso de trabajar con diferentes colaboradores, permite mantener ese historial de **quién** realizó los ajustes a los programas.

Por otro lado, también permite trabajar con diferentes versiones para poder unirlas luego. Esto permite poder trabajar con versiones pasadas y crear nuevas líneas de código para luego unirlas y tener así los siguientes pasos de nuestro proyecto.
